```python
import pandas as pd
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.tsa.stattools import acf, pacf

pd.options.display.float_format  = '{:.2f}'.format
```

```python
df = pd.read_csv('SPY.csv')
df.head()
```

```python
residuals = df['Adj Close'].pct_change() - df['Adj Close'].pct_change().mean()
residuals = residuals.dropna()
```

```python
acorr_ljungbox(residuals, boxpierce=True, return_df=True)
```

```python
# ljung-box from scratch
n = len(residuals)
c = n*(n+2)
racf = acf(residuals)[1:]
sum(map(lambda i: c*i[1]**2/(n-(i[0]+1)), enumerate(racf)))
```

```python
# box-pierce from scratch
n = len(residuals)

racf = acf(residuals)[1:]
sum(map(lambda i: n*i**2, racf))
```
