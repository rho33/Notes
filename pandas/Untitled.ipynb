```python
import pandas as pd
```

# View Data

```python
df = pd.read_csv('top50.csv', encoding='iso-8859-1', index_col=0)
# view first n rows
df.head(7)
```

```python
# view last n rows
df.tail(2)
```

```python
# view single column (Series object)
df['Genre']
```

```python
# view multiple columns (df object)
df[['Genre', 'Energy']]
```

```python
# view specific row by row num
df.iloc[9]
```

```python
# view specific row by row name (index label)
df.loc[10] # row is named 10 in this case
```

```python
# view slice of df
df.iloc[slice(20, 25)] # or df.iloc[20:25]
```

```python
#Note that you can also use a slice object to select rows without iloc function 
print(df[slice(20, 21)] )
# but using iloc by default for selecting rows seems to be better. 

#This is because df.iloc allows selecting rows and columns
print(df.iloc[1, 3])

# and it allows assigning values to cells.
df.iloc[0, 0] = 'changed value'
df.head(1)
```

```python
# filter based on column value
df[df['Genre']=='latin']
```

```python
# access cells by row and col numbers with df.iloc, or by column/
print(df.iloc[0, 2])

# or by row and column labels with df.loc 
print(df.loc[1, 'Genre'])
```

```python
df.iloc[10:20]
```
