```r
# petal width histograms
options(repr.plot.width=15, repr.plot.height=9)
par(mfrow = c(3, 1))
hist(
    iris$Petal.Width [iris$Species == 'setosa'],
    xlim = c(0, 3),
    breaks = 9,
    main = 'Petal Width for Setosa',
    xlab = '',
    col = 'red',
    cex.main = 2.5,
    cex.axis = 1.5,
    cex.lab = 1.75
)
hist(
    iris$Petal.Width [iris$Species == 'versicolor'],
    xlim = c(0, 3),
    breaks = 9,
    main = 'Petal Width for Versicolor',
    xlab = '',
    col = 'purple',
    cex.main = 2.5,
    cex.axis = 1.5,
    cex.lab = 1.75
)
hist(
    iris$Petal.Width [iris$Species == 'virginica'],
    xlim = c(0, 3),
    breaks = 9,
    main = 'Petal Width for Virginica',
    xlab = '',
    col = 'blue',
    cex.main = 2.5,
    cex.axis = 1.5,
    cex.lab = 1.75
)
```
